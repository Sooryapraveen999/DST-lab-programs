```python
class Node:
  def __init__(self,data):
    self.data=data
    self.next=None

class stack:
  def __init__(self):
    self.top=None
  #push operation
  def push(self):
    x=int(input("enter element to be inserted into stack"))
    new=Node(x)
    if self.top is None:
      self.top=new
      self.top.next=None
    else:
      new.next=self.top
      self.top=new
   #POP operation
  def pop(self):
    if self.top is None:
      print("STACK IS EMPTY")
    elif self.top.next is None:
      print("Popped element is :",self.top.data)
      print("--------------------------------")
      self.top=None
    else:
      temp=self.top
      print("Popped element is :",self.top.data)
      print("---------------------------------")
      self.top=temp.next
      temp=None
  def display(self):
    if self.top is None:
      print("stack is empty")
    else:
      temp=self.top
      temp=temp.next
    print("Top of the stack is ",self.top.data)
s=stack()
while(True):
  print("enter the choice to perform")
  print("1.push operation\n 2.pop operation\n 3.display operation\n 4.exit")
  option=int(input())
  if option==1:
    s.push()
  elif option==2:
    s.pop()
  elif option==3:
    s.display()
  elif option==4:
    print("You have exited from the stack")
    break
  else:
    print("enter valid option")
```
